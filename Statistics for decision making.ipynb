```python
# Task 1 - perform t-Test
from scipy import stats

altona_df = df[df['Suburb'] == 'Altona']
altona_prices = altona_df['Price'].dropna()
t_stat, p_value = stats.ttest_1samp(altona_prices, 800000)

mean_price = altona_prices.mean()
t_stat, p_value, mean_price

```

```python
# Task 2  - Compare Property Prices in Summer and Winter Months for the Year 2016

# to reform task 2 we need to check data type
df.info()
```

```python
# Lets convert date column to date time formate

df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df.info()
```

```python
# filter data for year 2016
df_2016 = df[df['Date'].dt.year == 2016]
df_2016
```

```python
# Define months for summer & winter season
months_summer = [4,5,6,7,8,9]
months_winter = [10,11,12,1,2,3]

#filtering data with respective months of season
summer_prices = df_2016[df_2016['Date'].dt.month.isin(months_summer)]['Price'].dropna()
winter_prices = df_2016[df_2016['Date'].dt.month.isin(months_winter)]['Price'].dropna()

# Lets perform 2 sample test

t_stat, p_value = stats.ttest_ind(summer_prices, winter_prices)

t_stat, p_value, summer_prices.mean(), winter_prices.mean(), len(summer_prices), len(winter_prices)
```

```python
# t-stat: -4.043
# p-value: 5.3309767667631686e-05
# mean property price in summer months: $1048054.7286917741
# mean property price in winter months: $1116647.5917391304
# number of summer month samples: 4036
# number of winter month samples: 2300

# With a p-value of 5.3309767667631686e-05, which is much less than the significance level of 0.05, we reject the null hypothesis. Therefore, we conclude that there is a significant difference in property prices between summer and winter months in 2016, with winter prices being higher on average.
```

```python
# Task 3 - Probability that out of 10 properties sold in Abbotsford, 3 will not have car parking

# Filter data for abbotsford 
abbotsford_df = df[df['Suburb'] == 'Abbotsford']
car_parking_counts = abbotsford_df['Car'].value_counts()
no_car_parking_count = car_parking_counts.get(0, 0)
total_properties = len(abbotsford_df)
total_properties


```

```python
# Calculation of probability of a property not having car parking
p_no_car_parking = no_car_parking_count / total_properties
```

```python
n = 10  # Number of properties sold
k = 3  # Number of properties without car parking
from scipy.stats import binom
prob = binom.pmf(k, n, p_no_car_parking)
prob = round(prob, 3)
print(f"The probability is: {prob}")
```

```python
# Task 4 - Probability finding a property with 3 rooms
abbotsford_df = df[df['Suburb'] == 'Abbotsford']
total_properties = len(abbotsford_df)
properties_with_3_rooms = (abbotsford_df['Rooms']== 3).sum()

probability = properties_with_3_rooms / total_properties
probability_rounded = round(probability, 3)

print(f"The probability is {probability_rounded}")
```

```python
# Task 4 - Probability finding a property with 2 bathrooms
abbotsford_df = df[df['Suburb'] == 'Abbotsford']
total_properties = len(abbotsford_df)

properties_with_2_bathrooms = (abbotsford_df['Bathroom']== 2).sum()
probability = properties_with_2_bathrooms / total_properties
probability_rounded = round(probability, 3)

print(f"The probability is {probability_rounded}")
```
